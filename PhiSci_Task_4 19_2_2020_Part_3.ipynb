### fisher exact

```python
from scipy.stats import fisher_exact
from statsmodels.stats.multitest import multipletests
```

```python
# Defining the sizes (number of genes) in clusters & background
clusters = [uniprot_cluster0, uniprot_cluster1]
clusters_size = [cluster.shape[0] for cluster in clusters]
cluster_label_size = dict(enumerate(clusters_size))
background_size = uniprot_background.shape[0]
```

```python
def p_value(background_count, cluster_count, cluster_label):
    cluster_size = cluster_label_size[cluster_label]
    
    # declaring fisher variables 
    corrected_cluster_term = cluster_count
    not_cluster_but_term = background_count
    cluster_but_not_term = cluster_size - cluster_count
    not_cluster_not_term = background_size - (corrected_cluster_term + background_count + cluster_but_not_term)
  
    # building Fisher matrix
    array = np.array([[corrected_cluster_term, not_cluster_but_term],
                      [cluster_but_not_term,   not_cluster_not_term]])
  
    # calling fisher_exact using fisher two-sided
    odds_ratio, p_value = fisher_exact(array, alternative='two-sided')
    
    return p_value
```

```python
# calculating the p values and modify it using multipletests for the cluster 0
for enrch_df in cluster_0:
    enrch_df['p_value'] = enrch_df.apply(lambda df: p_value(df.background, df.cluster0, 0), axis=1)
    enrch_df['modified_p_value'] = multipletests(enrch_df['p_value'], method='fdr_bh')[1]
```

```python
# calculating the p values and modify it using multipletests for the cluster 1
for enrch_df in cluster_1:
    enrch_df['p_value'] = enrch_df.apply(lambda df: p_value(df.background, df.cluster1, 1), axis=1)
    enrch_df['modified_p_value'] = multipletests(enrch_df['p_value'],method='fdr_bh')[1]
```
