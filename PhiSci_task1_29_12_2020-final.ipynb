```python
import os # for the directory 
import csv # reading files
import pandas as pd  # data analysis tools
import numpy as np
from pandas import DataFrame # easier to analysis
fname=r"C:\Users\mohab\gdc_download_PhiSci\text files" # locate the location 
os.chdir(r"C:\Users\mohab\gdc_download_PhiSci\text files") # assign the default work folder 
list_names = os.listdir(fname) # to include all files into one list variable
```

```python
# to assign the name of the first column
genes=[] 
#for file in list_names:
with open(list_names[0], newline = '\n') as fileByFile:
    fileByFileReader = csv.reader(fileByFile, delimiter='\t') # reading the file based on a specific separator format
    for fileByFile in fileByFileReader: # read all lines one by one
        genes.append(fileByFile[0]) # append line by line to the list

# the final table
finalTable = DataFrame (genes,columns=['gene_type'])
```

```python
gene_id = finalTable['gene_type']
for i in range(len(gene_id)):
    gene_str = str(gene_id[i])
    finalTable.loc[i]= gene_str[:gene_str.index('.')]
```

```python
finalTable.to_csv(r'Final_gene_id.csv', index = False)
```

```python
# create all columns to the final list
result=[] # creat a list to include all columns
for file in list_names[:-1]: # 
    with open(file, newline = '\n') as fileByFile:
        fileName = file[0:len(file)-9]
        rowByRowReader = csv.reader(fileByFile, delimiter='\t')
        for rowByRow in rowByRowReader:
            result.append(rowByRow[1])
    finalTable[fileName] = DataFrame(result, columns=[fileName])
    result=[]
```

```python
import json 
obj = open(r"C:/Users/mohab/data.json",) 
data = json.load(obj) 
cols = finalTable.columns
cols = cols.tolist()
len(cols) # 595
for i in range(0, 594):
    file_name = data[i]["submitter_id"]
    file_name = file_name[:-5]
    #index = cols.index(file_name)
    full_submit = data[i]["associated_entities"][0]["entity_submitter_id"]
    gene_id = full_submit[:-12]
    finalTable.rename(columns={file_name: gene_id}, inplace=True)
```
